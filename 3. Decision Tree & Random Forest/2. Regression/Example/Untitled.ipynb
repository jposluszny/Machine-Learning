```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('autos.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
cats = ['symboling', 'make', 'fuel-type', 'aspiration', 'num-of-doors',
       'body-style', 'drive-wheels', 'engine-location','engine-type', 
        'num-of-cylinders','engine-size', 'fuel-system']
```

```python
dfd = pd.get_dummies(df, columns=cats, drop_first = True)
```

```python
dfd.head()
```

```python
X = dfd.drop('price', axis=1)
y = dfd.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tr = DecisionTreeRegressor()
```

```python
tr.fit(X_train, y_train)
```

```python
predictions = tr.predict(X_test)
predictions
```

```python
ax = sns.scatterplot(x=y_test, y=predictions)
ax.set_ylabel('Predictions')
ax.set_xlabel('Price')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
```

```python
mean_absolute_error(y_test, predictions)
```

```python
mean_squared_error(y_test, predictions)
```

```python
root_mean_squared_error(y_test, predictions)
```

```python
sns.histplot((y_test - predictions), bins=50, kde=True)
```

```python
tr_import = tr.feature_importances_
importances = pd.DataFrame(tr_import, index=X_train.columns, columns=['Importance']).sort_values(by='Importance', ascending=False)
importances
```

```python
ax = sns.barplot(x=importances.head().Importance, y=importances.head().index)
ax.set_ylabel('Feature')
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rr = RandomForestRegressor()
```

```python
rr.fit(X_train, y_train)
```

```python
predictions = rr.predict(X_test)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
mean_squared_error(y_test, predictions)
```

```python
root_mean_squared_error(y_test, predictions)
```

```python
ax = sns.scatterplot(x=y_test, y=predictions)
ax.set_xlabel('Price')
ax.set_ylabel('Predictions')
```

```python
sns.histplot((y_test - predictions), kde=True)
```

```python
importances = rr.feature_importances_
```

```python
importances_df = pd.DataFrame(importances, index=X_train.columns, columns=['Importance']).sort_values(by='Importance', ascending=False)
```

```python
importances_df
```

```python
ax = sns.barplot(x=importances_df.head(10).Importance, y=importances_df.head(10).index)
ax.set_ylabel('Feature')
```
